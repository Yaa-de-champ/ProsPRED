```python
import SimpleITK as sitk
import numpy as np

# Load the DICOM image
dicom_path = r"C:\Users\Lenovo\Downloads\PROSTATE CANCER FILES\PROSTATEx\t2.dcm"
image = sitk.ReadImage(dicom_path)

# Convert the SimpleITK image to a NumPy array
image_array = sitk.GetArrayFromImage(image)

# Define the lesion coordinate as the center point
lesion_coordinate = (178, 140, 0)

# Define the size of the ROI
roi_size = (100, 100, 5)  # Example size

# Generate random coordinates around the lesion coordinate
x_min = max(0, lesion_coordinate[0] - roi_size[0] // 2)
x_max = min(image_array.shape[0], lesion_coordinate[0] + roi_size[0] // 2)
y_min = max(0, lesion_coordinate[1] - roi_size[1] // 2)
y_max = min(image_array.shape[1], lesion_coordinate[1] + roi_size[1] // 2)
z_min = max(0, lesion_coordinate[2] - roi_size[2] // 2)
z_max = min(image_array.shape[2], lesion_coordinate[2] + roi_size[2] // 2)

# Crop the image based on the ROI boundaries
cropped_array = image_array[x_min:x_max, y_min:y_max, z_min:z_max]

# Convert the NumPy array back to a SimpleITK image
cropped_image = sitk.GetImageFromArray(cropped_array)

# Set the image spacing and origin
cropped_image.SetSpacing(image.GetSpacing())

cropped_image.SetOrigin(image.TransformIndexToPhysicalPoint((x_min, y_min, z_min)))

# Save the cropped image
output_path = r"C:\Users\Lenovo\Downloads\PROSTATE CANCER FILES\PROSTATEx\t2cropped_image.dcm"

sitk.WriteImage(cropped_image, output_path)

# Print the shape of the cropped image
print("Shape of the cropped image:", cropped_image.GetSize())
```
